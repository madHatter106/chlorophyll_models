```python
def make_tab(df_, cat_list, biom_str):
    def make_new_source(cat_list):
        filtered_df = df_[df_[selection_label].isin(cat_list)]
        return ColumnDataSource(filtered_df)

    def make_checkboxgroup(category_list):
        cbg_selection = CheckboxGroup(labels=category_list,
                                     active=[0],
                                     sizing_mode='scale_width')
        return cbg_selection

    def update(attr, old, new):
        # get the list of selections to graph
        selections_to_plot = [selection.labels[i] for i in selection.active]

        # make a new dataset based on the selection and the make_dataset function
        new_source = make_new_source(selections_to_plot)
        # Update the source used in the plot
        source_.data.update(new_source.data)
        
    def make_plot(src):
        #biome_selection = CheckboxGroup(labels=df.biome.unique().tolist(), )
        p = figure(plot_width=1000, plot_height=1000, title=plot_title,
                   x_axis_label=x_lbl, y_axis_label=y_lbl,
                  x_range=[-0.43, 1.3], y_range=[-2, 2])
        p.circle(x=x_data_col_name, y=y_data_col_name, fill_color='sst_color', fill_alpha=0.4,
                 size='etopo_l_adj', line_color='black',source=src,)
        hover = HoverTool(tooltips=cur_tool_tip)
        p.add_tools(hover)
        return p
    
    selection = make_checkboxgroup(cat_list)
    selection.on_change('active', update)
    initial_selection = [selection.labels[i] for i in selection.active]
    source_ = make_new_source(initial_selection)
    cur_tool_tip = [('SST', '@oisst'), ('depth(m)', '@depth)')]
    p = make_plot(source_)
    control=WidgetBox(selection, width=200, sizing_mode='scale_both')
    # create a row layout
    layout = row(control, p)
    # make a tab
    tab = Panel(child=layout,title=biom_str)
    return tab


def modify_category_plot2(doc):
    def get_source_n_cats(biome):
        df_biome = dfs.loc[dfs.biome==biome, data_name_list]
        cat_list = df_biome[selection_label].unique().tolist()
        cat_list.sort()
        n_cat_list = len(cat_list)
        color_list = viridis(n_cat_list)
        df_biome.assign(tab_color=NaN)
        #for cat, color in zip(cat_list, color_list):
        #    df_biome.loc[df_biome[selection_label] == cat, 'tab_color'] = color
        return df_biome, cat_list
    
    data_name_list = [x_data_col_name, y_data_col_name, selection_label, 'mxBlId', 'provCode']
    df_coast, coastal_list = get_source_n_cats('Coastal')
    df_polar, polar_list = get_source_n_cats('Polar')
    df_trades, trades_list = get_source_n_cats('Trades')
    df_west, west_list = get_source_n_cats('Westerlies')
    tab_coast = make_tab(df_coast, coastal_list, 'Coastal')
    tab_polar = make_tab(df_polar, polar_list, 'Polar')
    tab_trades = make_tab(df_trades, trades_list, 'Trades')
    tab_west = make_tab(df_west, west_list, 'Westerlies')
    tabs = Tabs(tabs=[tab_coast, tab_polar, tab_trades, tab_west])
    doc.add_root(tabs)
    return doc

def modify_category_plot(doc):
    def get_source_n_cats(biome):
        df_biome = dfs.loc[dfs.biome==biome, data_name_list]
        cat_list = df_biome[selection_label].unique().tolist()
        cat_list.sort()
        n_cat_list = len(cat_list)
        color_list = viridis(n_cat_list)
        df_biome.assign(tab_color=NaN)
        #for cat, color in zip(cat_list, color_list):
        #    df_biome.loc[df_biome[selection_label] == cat, 'tab_color'] = color
        return df_biome, cat_list
    
    data_name_list = [x_data_col_name, y_data_col_name, selection_label, 'mxBlId', 'provCode',
                      'oisst', 'sst_color', 'etopo_l_adj', 'depth']
    df_coast, coastal_list = get_source_n_cats('Coastal')
    df_polar, polar_list = get_source_n_cats('Polar')
    df_trades, trades_list = get_source_n_cats('Trades')
    df_west, west_list = get_source_n_cats('Westerlies')
    tab_coast = make_tab(df_coast, coastal_list, 'Coastal')
    tab_polar = make_tab(df_polar, polar_list, 'Polar')
    tab_trades = make_tab(df_trades, trades_list, 'Trades')
    tab_west = make_tab(df_west, west_list, 'Westerlies')
    tabs = Tabs(tabs=[tab_coast, tab_polar, tab_trades, tab_west])
    doc.add_root(tabs)
```
